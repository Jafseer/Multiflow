# Load data from a file as a stream and save test results into a file.

### Load the data set as a stream

For this purpose we will use the **FileStream** class:

```python
%matplotlib notebook
```

```python
from skmultiflow.data.file_stream import FileStream
```

```python
from skmultiflow.data import FileStream
```

```python
stream = FileStream("D:/Streaming data set/streaming-datasets-master/airlines.csv")
#D:\Streaming data set\streaming-datasets-master\airlines.csv
```

### Instantiate the Hoeffding Tree classifier

We will use the default parameters.

```python
from skmultiflow.trees import HoeffdingTreeClassifier
```

```python
ht = HoeffdingTreeClassifier()
```

### Setup the evaluator

we will use the **EvaluatePrequential** class.

```python
from skmultiflow.evaluation import EvaluatePrequential
```

```python
evaluator = EvaluatePrequential(show_plot=True,pretrain_size=1000,
                               max_samples=10000,output_file='airline_results.csv')
```

### Run the evaluation

```python
evaluator.evaluate(stream=stream, model=ht)
```
